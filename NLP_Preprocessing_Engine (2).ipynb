# Task 1: Conceptual Understanding

1. Difference between "Love" and "love" in NLP

"Love" and "love" are treated as different tokens in case-sensitive NLP systems.
Lowercasing ensures both are treated as the same word.


2. What happens if stopwords are not removed?

If stopwords are not removed, they increase noise in the dataset and reduce model performance because frequent words like "is", "the", and "and" dominate the text.


3. Two real-world scenarios where removing stopwords can be harmful

Example 1:
"I am not happy"
Removing "not" changes sentiment meaning.

Example 2:
"To be or not to be"
Stopwords are necessary to preserve sentence meaning.


4. Difference between stemming and lemmatization

Stemming removes suffixes without considering grammar.
Example:
playing → play

Lemmatization converts words to dictionary base form.
Example:
better → good

# Task 3: Build Advanced Preprocessing Function

Objective:

Create a preprocessing function that cleans noisy real-world text.

Steps performed in preprocessing:

• Remove URLs
• Remove email patterns
• Remove numbers
• Normalize repeated characters
• Convert text to lowercase
• Remove special characters
• Remove extra spaces
• Remove very short tokens (except "no" and "not")

Function name:

preprocess_text(text)

```python
# Task 3: Preprocessing Function Implementation

def preprocess_text(text):

    # Handle empty or invalid input
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove email-like patterns
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Normalize repeated characters (loooove → love)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Convert to lowercase
    text = text.lower()

    # Remove special characters
    text = re.sub(r"[^a-z\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenize text
    tokens = text.split()

    # Remove very short tokens except "no" and "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    # Join tokens back into cleaned sentence
    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence
```

# Task 4: Stress Testing the Preprocessing Function

Objective:

Test preprocessing function on 10 diverse noisy sentences containing:

• Emojis
• Numbers
• URLs
• Repeated characters
• Mixed uppercase text
• Symbols
• Slang

For each sentence display:

Original Text
Cleaned Tokens
Cleaned Sentence

```python
# Task 4: Stress Testing Input Data

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
# Task 4: Apply Preprocessing

processed_results = []

for sentence in test_sentences:

    tokens, clean_sentence = preprocess_text(sentence)

    processed_results.append(tokens)

    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Clean sentence:", clean_sentence)
    print("-" * 40)
```

# Task 5: Token Analytics

Objective:

Perform token-level statistical analysis after preprocessing.

Metrics calculated:

• Total number of tokens
• Number of unique tokens
• Average token length

```python
# Task 5: Token Analytics Calculation

for tokens in processed_results:

    if len(tokens) == 0:
        continue

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_token_length = sum(len(word) for word in tokens) / total_tokens

    print("Tokens:", tokens)
    print("Total tokens:", total_tokens)
    print("Unique tokens:", unique_tokens)
    print("Average token length:", avg_token_length)
    print("-" * 40)
```

# Task 6: Frequency Analysis

Objective:

Combine tokens from all processed sentences and calculate word frequency.

Using:

collections.Counter

Outputs:

Top 10 most frequent words
Top 5 least frequent words

```python
# Task 6: Frequency Analysis

all_tokens = []

for tokens in processed_results:
    all_tokens.extend(tokens)

word_frequency = Counter(all_tokens)

print("Top 10 frequent words:")
print(word_frequency.most_common(10))

print("\nTop 5 least frequent words:")
print(word_frequency.most_common()[-5:])
```

# Task 7: Build Full NLP Pipeline

Objective:

Create reusable function full_pipeline(text_list)

Returns:

• Combined tokens
• Cleaned sentences

This allows batch preprocessing of multiple sentences efficiently.

```python
# Task 7: Full Pipeline Function

def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:

        tokens, sentence = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
# Task 7: Pipeline Testing

pipeline_result = full_pipeline(test_sentences)

print(pipeline_result)
```
