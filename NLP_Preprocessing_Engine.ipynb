Difference between "Love" and "love" in NLP

They are treated as different tokens in case-sensitive NLP. Converting to lowercase ensures both become the same word.

What happens if stopwords are not removed?

Stopwords increase noise and reduce model performance.

Two scenarios where removing stopwords is harmful

1. Sentiment Analysis
Example: "I am not happy" removing "not" changes meaning

2. Question answering systems
Example: "What is AI?" removing "what" affects interpretation

Difference between stemming and lemmatization

Stemming removes suffixes (running → run)
Lemmatization converts words into dictionary base form (better → good)

```python
#TASK2(MAIN FUNCTION)
import re

def preprocess_text(text):

    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)
    text = re.sub(r"\d+", "", text)

    text = re.sub(r"(.)\1{2,}", r"\1", text)

    text = text.lower()

    text = re.sub(r"[^a-z\s]", "", text)

    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()

    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence
```

```python
#TASK3(STRESS TESTING)

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    results.append((sentence, tokens, cleaned))

for r in results:
    print("\nOriginal:", r[0])
    print("Tokens:", r[1])
    print("Cleaned Sentence:", r[2])
```

```python
#TASK4(TOKEN ANALYTICS)

analytics = []

for original, tokens, cleaned in results:

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_token_length = np.mean([len(t) for t in tokens]) if tokens else 0

    analytics.append((original, total_tokens, unique_tokens, avg_token_length))

for a in analytics:
    print("\nSentence:", a[0])
    print("Total Tokens:", a[1])
    print("Unique Tokens:", a[2])
    print("Average Token Length:", round(a[3],2))
```

```python
#TASK5(FREQUENCY ANALYSIS)
all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

freq = Counter(all_tokens)

print("Top 10 frequent words:")
print(freq.most_common(10))

print("\nTop 5 least frequent words:")
print(freq.most_common()[-5:])
```

```python
#TASK6(FULL PIPELINE FUNCTION)
def full_pipeline(text_list):

    tokens_list = []
    cleaned_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        tokens_list.append(tokens)
        cleaned_sentences.append(cleaned)

    return {
        "tokens": tokens_list,
        "clean_sentences": cleaned_sentences
    }

pipeline_output = full_pipeline(test_sentences)

pipeline_output
```

```python
#TASK7(ERROR HANDLING)
edge_cases = ["", "😂😂😂😂", "123456"]

for case in edge_cases:

    tokens, cleaned = preprocess_text(case)

    print("\nInput:", case)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
```
